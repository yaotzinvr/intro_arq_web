```python
def ejecuta(comando,commit=False):
    conn=None
    cur=None
    try:
        conn =  psycopg2.connect(cadena_conexion)
        cur = conn.cursor()
        cur.execute(comando)
        if commit: conn.commit() 
        res = cur.fetchall()
    except Exception as e:
        return str(e)
    finally:
        if cur: cur.close()
        if conn: conn.close()
    return res
        
```

# Importando Pokemones - Crea BD

```python
df = pd.read_csv('datasets_2756_4568_pokemon.csv')
```

```python
df
```

```python
script_create_table = """
CREATE TABLE {table_name} (
            id SERIAL PRIMARY KEY,
            {columnas_definition}
        )
"""
```

```python
cols_create = []
for c in df.columns:
    n = max(df[c].map(lambda x: len(str(x))))
    cols_create.append("{} VARCHAR({}) NULL".format(c,n))
```

```python
cols_create
```

```python
','.join(cols_create)
```

```python
script_create_table.format(**{'table_name':'pokemon','columnas_definition':','.join(cols_create)})
```

```python
comando = script_create_table.format(**{'table_name':'pokemon','columnas_definition':','.join(cols_create)})
ejecuta(comando,commit=True)
```
